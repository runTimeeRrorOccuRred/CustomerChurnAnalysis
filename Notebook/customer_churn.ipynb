```python
# plotting barchart to see the proportion of churn rate affected by the categorical columns.
fig,axes=plt.subplots(nrows=8,ncols=2,figsize=(16,40))
axes=axes.flatten()
for i,col in enumerate(cat_col):
    churn_rate=pd.crosstab(cat_df[col],cat_df['Churn'],normalize='index')*100
    axx=churn_rate.plot(kind='bar',color=['skyblue','salmon'],ax=axes[i])
    axes[i].set_title(f'Churn Rate by {col}')
    axes[i].set_xlabel('')
    axes[i].set_ylabel('Churn Rate')
    axes[i].set_xticklabels(axes[i].get_xticklabels(),rotation=0)
    for container in axx.containers:
        axx.bar_label(container)
plt.show()
```

* Older customers most likely to leave the service rather than younger customers. It may be because of affordability for being more price sensitive or not having tech support make them leave the service.
* Customer having partner are less likely to churn than not having partner.One possible explanation is that customers with partners may share services (like TV, internet, and phone), which increases switching costs and reduces churn.
* Customers having dependent are less likely to leave the service.
* Customers having fiber optic internet service are more likely to leave the service rather than other internet service where customers not having internet service very less likely to leave the service. Possible reason behind that may be those customers having fiber optic internet servie are suffering low quality or slow internet service where as those customers not having internet service they are opted for phone service with lower cost with fewer complications.
* Customers who do not have online security, online backup, device protection, and tech support are more likely to leave the service. A possible solution is to enhance these add-on services to provide better protection,support and security, thereby improving customer satisfaction and retention.
* Churn is higher among customers with internet services like using streaming Tv or Movies around 30% compared to those without internet (7%), suggesting internet-related factors may drive churn possibly due to service performance, pricing, or dependency on internet connectivity.
* Customers with month-to-month contracts have a much higher churn rate (42%) compared to those with one-year (11%) and two-year (2%) contracts. This indicates that customers on shorter contracts are more likely to discontinue the service, while long-term contracts help retain customers by increasing commitment and reducing the likelihood of switching.
* Customers who use paperless billing have a higher churn rate (33.5%) compared to those who receive paper bills (16.3%).
This suggests that customers on paperless billing may often be on month-to-month or online-only plans, which provide greater flexibility to cancel services.
Alternatively, it could indicate that paper billing customers are more traditional, loyal, or on long-term contracts.
* Customers having the payment method electronic check are more likely to leave the service.Electronic check users might face payment failures, delays, or less convenience compared to automatic credit card or bank payments.Electronic check payments rely on a stable internet connection.
Customers with poor or unreliable internet might face failed or delayed payments, leading to billing issues or service suspension, which increases dissatisfaction which leads them to leave the service.

```python
# churn rate by multiple services
def multiService(row):
    if row['PhoneService']=='Yes' and row['InternetService'] in ('DSL','Fiber optic'):
        return 'Both'
    elif row['PhoneService']=='Yes' and row['InternetService']=='No':
        return 'PhoneService'
    elif row['PhoneService']=='No' and row['InternetService'] in ('DSL','Fiber optic'):
        return 'InternetService'
    else:
        None
final_data['MultipleService']=final_data.apply(multiService,axis=1)
    
    
```

```python
churnMulti=final_data.groupby('MultipleService')['Churn'].value_counts(normalize=True).mul(100).rename('ChurnRate').reset_index()
```

```python
churnMulti=churnMulti[churnMulti['Churn']=='Yes']
```

```python
plt.figure(figsize=(8,5))
ax=sns.barplot(data=churnMulti,x='MultipleService',y='ChurnRate',palette='viridis')
for container in ax.containers:
    ax.bar_label(container)
ax.set_title('Multiple Service with ChurnRate')
plt.savefig('multiple_service_churn_rate.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# year-over-year growth of customers by tenure.
tenure_data=final_data.groupby('tenure')['customerID'].count().rename('CurrentCustomers').reset_index()
```

```python
tenure_data['PreviousCustomers']=tenure_data['CurrentCustomers'].shift(1)
```

```python
tenure_data.PreviousCustomers.fillna(0,inplace=True)
```

```python
tenure_data['PreviousCustomers']=tenure_data['PreviousCustomers'].astype(int)
```

```python
tenure_data['TenureWiseCustomers']=(tenure_data['CurrentCustomers']-tenure_data['PreviousCustomers'])*100/tenure_data['PreviousCustomers']
```

```python
tenure_data
```

```python
tenure_data['TenureWiseCustomers'] = tenure_data['TenureWiseCustomers'].clip(upper=200)
plt.figure(figsize=(8,5))
plt.plot(tenure_data['tenure'],tenure_data['TenureWiseCustomers'])
# Add labels and title
plt.title('Tenure-Based Customer Growth Rate', fontsize=14, fontweight='bold')
plt.xlabel('Tenure (Months)', fontsize=12)
plt.ylabel('Growth Rate (%)', fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```
